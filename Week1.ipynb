```python
import csv

%precision 2 #this is a IPython module
with open('mpg.csv') as csvfile:
    mpg= list(csv.DictReader(csvfile))
    
mpg[:3]
```

```python
mpg[0].keys()
```

```python
#very clean way to obtain sum
sum (float(d['weight']) for d in mpg) / len(mpg)
```

```python
#set gives the unique values of the list
cylinders = set (d['cylinders'] for d in mpg)
cylinders
```

```python
HpMpgByCly =[]
for c in cylinders:
    summpg=0;
    cyltypecount=0
    for d in mpg:
        if d['cylinders'] == c:
            summpg += float(d['weight'])
            cyltypecount +=1
    HpMpgByCly.append((c, summpg/cyltypecount)) #it has to be 
    #(x,y) cause it create a list of tuples
    
            
HpMpgByCly.sort( key= lambda x : x[0]) #x is the tuple on HpMpgByCly , by choosing x[0] or x[1] select which
#element of the tuple will be ascending

HpMpgByCly
```

```python
# Now that we know how the python regex API works, lets talk about more complex patterns. The regex
# specification standard defines a markup language to describe patterns in text. Lets start with anchors.
# Anchors specify the start and/or the end of the string that you are trying to match. The caret character ^
# means start and the dollar sign character $ means end. If you put ^ before a string, it means that the text
# the regex processor retrieves must start with the string you specify. For ending, you have to put the $
# character after the string, it means that the text Regex retrieves must end with the string you specify.

# Here's an example
text = "Amy works diligently. Amy gets good grades. Our student Amy is succesful."

# Lets see if this begins with Amy
re.findall("^Amy",text)  #only one occurence, which is the first Amy.
```

# Patterns and Character Classes

```python
import re
grades="ACAAAABCBCBAAD"

print(re.findall('A',grades)) #counts every A 
print(re.findall('AB',grades)) # counts every AB in the string
print(re.findall('[AB]',grades)) #counts every A or B
print(re.findall('A|B',grades)) #counts every A or B
print(re.findall('A[B-D]',grades)) #counts every A followed by B to D
print(re.findall('[^A]',grades)) #counts every NOT A
```

# Quantifiers

```python
grades="ACAAAABCBCBAAD"

print(re.findall('A{2,10}',grades)) #finds A's thar occur 2 to 10 times together

# note that if you set :
print(re.findall('A{1,1}A{1,1}',grades)) 
# the code finds more AA than the previous version.


# It's important to note that the regex quantifier syntax does not allow you to deviate from the {m,n}
# pattern. In particular, if you have an extra space in between the braces you'll get an empty result
print(re.findall("A{2, 2}",grades))
```

```python
# Now, that's a bit of a hack, because we included a maximum that was just arbitrarily large. There are three
# other quantifiers that are used as short hand, an asterix * to match 0 or more times, a question mark ? to
# match one or more times, or a + plus sign to match one or more times. Lets look at a more complex example,
# and load some data scraped from wikipedia
with open("ferpa.txt","r") as file:
    # we'll read that into a variable called wiki
    wiki=file.read()
# and lets print that variable out to the screen
wiki
```

```python
re.findall('[a-zA-Z]{1,100}\[edit\]',wiki)
```

```python
re.findall('\w{1,100}\[edit\]',wiki) # \w mathes any character
```

```python
re.findall('\w*\[edit\]',wiki) # * mathes any size
```

```python
re.findall('[\w ]*\[edit\]',wiki) # * mathes any size and we added space on the  [] condition
```

```python
# now, if we want the tittles, we split removing the [edit]
for title in re.findall('[\w ]*\[edit\]',wiki):
    print(re.split('\[edit]',title)[0])
```

```python
re.findall('(\w*)(\[edit\])', wiki)
```

```python
for item in re.finditer('(?P<title>[\w ]+)(?=\[edit\])',wiki):

    print(item.groupdict())
```
